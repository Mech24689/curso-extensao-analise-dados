```python
#Aqui criamos um novo dataframe com as colunas que eu quero...
#Cria mais de uma coluna
df2 = df[['Revenda', 'Municipio', 'Produto', 'Valor de Venda', 'Bairro']]

display(df2)
```

```python
#loc -> localiza uma linha e coluna
#Exibe a 4a. linha
display(df2.loc[9:"JEQUIE"])
```

```python
#Criar um dataframe gas_df contendo
#apenas as 4 colunas (revenda, municipio, produto, valor de venda)
#somente com combustivel sendo gasolina e exibir na tela

gas_df = df2.loc[df['Produto'] == 'GASOLINA']
display(gas_df)

display(gas_df['Valor de Venda'].max())
```

```python
#Exibir qual é o posto e municipio da gasolina mais cara do país?
display(gas_df[['Revenda','Municipio', 'Valor de Venda']].max())
```

```python
#DataFrame.loc[] com múltiplas condições para filtragem
#Quais são os preços, postos que vendem ETANOL na minha cidade INDAIATUBA

df_etanol = df2.loc[(df2['Produto'] == 'ETANOL') & (df2['Municipio'] == 'INDAIATUBA')]
df_etanol.sort_values(by='Valor de Venda', inplace=True)

display(df_etanol)
```

```python
#Qual a média de  preços dos combustiveis GASOLINA e GASOLINA ADITIVADA do bairro BELA VISTA em SÃO PAULO

df_test = df[['Valor de Venda', 'Produto', 'Bairro', 'Municipio']]
df_test_filter = df_test.loc[(df_test['Municipio'] == 'SAO PAULO') & (df_test['Bairro'] == 'MOOCA') & (df_test['Produto'].isin(['GASOLINA','GASOLINA ADITIVADA']))]
display(df_test_filter)
```

```python
print(df_test_filter.mean())
```

```python
# Como mostrar média de valor de venda POR COMBUSTIVEL Brasil todo?
media_por_combustivel = df2[['Produto', 'Valor de Venda']].groupby(by='Produto').mean().round(2)

display(media_por_combustivel)
```

```python
#Vamos brincar de grafico
import matplotlib.pyplot as plt


plt.hist(df['Valor de Venda'])
#Vamos colocar um titulo no grafico
plt.title("Preço dos combustveis - Nov/2021")
#Rótulo horizontal e vertical
plt.xlabel('Preço (em reais)')
plt.ylabel("Quantidade de Coletas")

#traça a  linha vermelha tracejada com o preço médio
plt.axvline(df['Valor de Venda'].mean(), color='red', linestyle='dashed', linewidth='1')

#"plota" o gráfico
plt.show()
```

```python
c_mean = df['Valor de Venda'].groupby(by=df['Produto']).mean()
```

```python
#Vou definir a area do grafico
plt.figure(figsize=(10,5))
#plotar o grafico
c_mean.plot(
    kind="barh",
    xlabel='Tipo de Combustivel',
    ylabel='Preço reais/litro',
    color='red'
)
#grid
plt.grid()

#exibe
plt.show()
```

```python
plt.figure(figsize=(10,5))

c_mean.plot(kind="barh", 
    ylabel="Combustíveis", 
    xlabel="Preço/litro", 
    color="purple", 
    title="Média dos Preços/litro de cada produto", 
    alpha=0.5)
```

```python
display(c_mean)

excel = "Por_litro.xlsx"
c_mean.to_excel(excel, "Sumário")
```

```python
#vamos personalizar a planlha excel
from openpyxl import load_workbook
from openpyxl.styles import PatternFill, Font, Color, Alignment

#wb = Workbook
wb = load_workbook(excel)

#Pegar a planilha certa... usando o Sheet Name (nome da planilha)

#ws = WorkSheet/ planilha atual, ativa, de trabalho
ws = wb['Sumário']

#Vamos pintar o cabeçalho da tabela de "cinzinha"
cinzinha = PatternFill("solid", fgColor="808080")
coords = ['A1', 'B1']
for coord in coords:
  ws[coord].fill = cinzinha

#Onde o preço do combustivel for maior ou igual a 6,5 reais (6.5) pinta a fonte
#de vermelho e deixa negrito...
MAX_ROW = ws.max_row
num_linha = 2
while(num_linha <= MAX_ROW):
  coord = 'B'+str(num_linha)
  if ws[coord].value > 6.5:
    ws[coord].font = Font(bold=True, color="FF0000")
  num_linha = num_linha + 1

#salvar o excel
wb.save(excel)
```
